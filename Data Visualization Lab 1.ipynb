```python
__author__="Rajat Kulshreshtha, 15BCE0544"
import pandas as pd

dataset = pd.read_csv("C:/Users/RAJAT KULSHRESHTHA/fortune1000.csv")
```

```python
# Q1. Create a pandas DataFrame from the above file:
pd.DataFrame(dataset)
print(type(dataset))
```

```python
# Q2. Display first 5 rows of the Data:
dataset.head()
```

```python
# Q3. How many unique sectors are in the dataset
len(dataset["Sector"].unique())
```

```python
# Q4. How many unique industries are in dataset
len(dataset["Industry"].unique())
```

```python
# Q5. How many companies are in the same each sector:
dataset.Sector.value_counts().reset_index(name='Company')
```

```python
# Q6. How many companies are in each industry:
dataset.Industry.value_counts().reset_index(name='Company')
```

```python
#Q7 How many industries are in each sector
dataset.Sector.value_counts().reset_index(name='Industries')
```

```python
#Q8 What are the top 5 Sectors based on employee size. Show sectors and total employee number.
dfQues8 = pd.pivot_table(aggfunc=sum,index=['Sector'],values=['Employees'], data=dataset)
toShow = dfQues8.sort_values(by='Employees',ascending=False)
toShow.head()
```

```python
#Q9 What are the top 10 industries based on total revenue, Show industries and their revenues
dfQues9 = pd.pivot_table(aggfunc=sum, index=['Industry'], values=['Revenue'],data=dataset)
toShow2=dfQues9.sort_values(by='Revenue', ascending=False)
toShow2[:10]
```

```python
#Q10 List the top 20 companies based on their profits, show their sectors, industries and their profits
dfQues10 = pd.pivot_table(aggfunc=sum,index=['Company','Sector','Industry'],values=['Profits'], data=dataset)
dfQues10.sort_values(by='Profits',ascending=False)[:20]
```

```python
#Q11 List the average revenue, profit and employee size for each industry
dfQues11 = pd.pivot_table(aggfunc='mean', index=['Industry'], values=['Revenue','Profits','Employees'],data=dataset)
dfQues11
# Here we are listing average revenue, profit and 'average' employee size.
```

```python
#Q12 Group the companies by sector and industry, list their average revenue, profit and employee size.

dataset.groupby(["Industry","Sector"])["Industry","Revenue","Profits","Employees"].mean()
```

```python
#Q13 Find the top 10 industries based on average profit/revenue ratio
dataset["Profit Revenue Ratio"] = dataset["Profits"]/dataset["Revenue"]
dataset.groupby("Industry")["Industry","Profit Revenue Ratio"].mean().sort_values(by='Profit Revenue Ratio', ascending=False)[:10]
```

```python
#Q14 Find Top 10 Companies based on their Profits/Revenue Ratio:
dataset["Profit Revenue Ratio"] = dataset["Profits"]/dataset["Revenue"]
dataset.groupby("Company")["Company","Profit Revenue Ratio"].mean().sort_values(by='Profit Revenue Ratio', ascending=False)[:10]
```
